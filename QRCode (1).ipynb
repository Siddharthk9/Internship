```python
import cv2 
from pyzbar.pyzbar import decode
import numpy as np

def decode_qr(image):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Use pyzbar to decode the QR code
    decoded_objects = decode(gray)
    
    for obj in decoded_objects:
        # Draw the bounding box around the detected QR code
        points = obj.polygon
        if len(points) > 4:
            hull = cv2.convexHull(np.array(points, dtype=np.float32))
        else:
            hull = points

        hull = list(map(tuple, np.squeeze(hull)))
        n = len(hull)
        for j in range(0, n):
            cv2.line(image, hull[j], hull[(j + 1) % n], (0, 255, 0), 3)

        # Display the decoded QR code data
        print(f"Data: {obj.data.decode('utf-8')}")

    return image

def main():
    # Open the video capture
    cap = cv2.VideoCapture(0)

    while True:
        # Capture frame-by-frame
        ret, frame = cap.read()
        if not ret:
            break

        # Decode QR code in the frame
        frame = decode_qr(frame)

        # Display the resulting frame
        cv2.imshow('QR Code Scanner', frame)

        # Exit if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # When everything is done, release the capture
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
