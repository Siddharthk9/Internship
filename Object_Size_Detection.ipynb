```python
import cv2

# Initialize the video capture
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open video stream from webcam.")
    exit()

# Define initial zoom level
scale_factor = 1.0

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    # If frame is read correctly ret is True
    if not ret:
        print("Error: Cannot receive frame (stream end?). Exiting ...")
        break
    
    # Get the dimensions of the frame
    height, width, _ = frame.shape
    
    # Define the region of interest (ROI)
    x1 = int(width * (1 - scale_factor) / 2)
    y1 = int(height * (1 - scale_factor) / 2)
    x2 = int(width - width * (1 - scale_factor) / 2)
    y2 = int(height - height * (1 - scale_factor) / 2)
    
    # Crop the frame to the ROI
    zoomed_frame = frame[y1:y2, x1:x2]
    
    # Display the resulting frame
    cv2.imshow('Zoomed Webcam', zoomed_frame)
    
    # Exit the loop when 'q' key is pressed
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    elif key == ord('+'):  # Zoom in
        scale_factor *= 100
    elif key == ord('-') and scale_factor > 100:  # Zoom out (with a lower limit)
        scale_factor /= 100

# When everything is done, release the capture
cap.release()
cv2.destroyAllWindows()
```
